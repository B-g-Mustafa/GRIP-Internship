# Task # 4 To Explore Decision Tree Algorithm

# Importing necessary packages

```python
import pandas as pd
import seaborn as sea
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

# Exploring Data

```python
iris = pd.read_csv("Iris.csv")
print ("Total data points: ",iris.shape,"\n")
print ("Columns(features) in dataset: ",iris.columns,"\n")
print("Total data points for each features:\n",iris["Species"].value_counts())
```

```python
iris.head
```

# Scatter Plot

```python
sea.set_style("whitegrid");
sea.FacetGrid(iris, hue="Species", size=4).map(plt.scatter, "SepalLengthCm", "SepalWidthCm").add_legend();
plt.show();
```

#  Pair-plot

```python
plt.close();
sea.set_style("whitegrid");
sea.pairplot(iris, hue="Species", size=3);
plt.show()
```

# Applying Decision Tree Classifier

```python
x=iris.drop(columns=['Id','Species'])
y=iris.drop(columns=['Id', 'SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'])

from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(x,y)
```

# Visualizing Decision Tree Classifier

```python
from sklearn import tree
fig = plt.figure(figsize=(25,20))
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa', 'versicolor', 'virginica']
viz= tree.plot_tree(dtc,feature_names=fn,class_names=cn,filled=True)
fig.savefig('treeviz.png')
```
